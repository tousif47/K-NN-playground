```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as pyplot
import pandas as pd
import json
from datetime import datetime
from datetime import time

pyplot.rcParams["figure.figsize"] = (10, 5)
```

```python
with open("flow_rates.json", "r") as f:
    data = json.load(f)

# Length of data
len(data)
```

```python
temp = list()
for i in range(len(data)):
    if (data[i]['value'] != 0):
        data[i]['timestamp'] = datetime.fromtimestamp(data[i]['ts'])
        data[i]['time'] = data[i]['timestamp'].time()
        temp.append(data[i])

data = temp
```

```python
print(len(data))
data.sort(key=lambda x:x['timestamp'])
data[0:5]
```

```python
df = pd.DataFrame(data[450:600])
pyplot.figure()
pyplot.plot(df['timestamp'], df['value'])
pyplot.xlabel('Time')
pyplot.ylabel('Flow')
pyplot.suptitle("Test")
pyplot.show()
```
