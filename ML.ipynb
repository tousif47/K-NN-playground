```python
%matplotlib notebook
import math
import numpy as np
import matplotlib.pyplot as pyplot
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("apartment_testing_data.csv") 
data.head()
```

```python
data.drop(labels="Description", axis=1, inplace=True)
data.drop(labels="ID", axis=1, inplace=True)
data.drop(labels="Event/Action", axis=1, inplace=True)
data.drop(labels="Devices F.#1", axis=1, inplace=True)
data.drop(labels="Devices F.#2", axis=1, inplace=True)
data.drop(labels="Devices F.#3", axis=1, inplace=True)
data.drop(labels="Tap %", axis=1, inplace=True)
data.drop(labels="Apt. #", axis=1, inplace=True)
data.drop(labels="Fixture #1", axis=1, inplace=True)
data.drop(labels="Fixture #2", axis=1, inplace=True)
data.drop(labels="Fixture #3", axis=1, inplace=True)
data.drop(labels="Result", axis=1, inplace=True)
data.drop(labels="Notes", axis=1, inplace=True)
data.drop(labels="Joel's Remarks", axis=1, inplace=True)
data.drop(labels="Start Time", axis=1, inplace=True)
data.drop(labels="End Time", axis=1, inplace=True)
data.drop(labels="Start (datetime)", axis=1, inplace=True)
data.drop(labels="Stop (datetime)", axis=1, inplace=True)
data.drop(labels="Duration", axis=1, inplace=True)
data.drop(labels="Duration.1", axis=1, inplace=True)
data.head()
```

```python
# In Order From 0: Kitchen Sink, Bathroom Sink, Bathroom Shower, Bathroom Toilet
y = data["Fixture Code"].to_numpy()
```

```python
data.drop(labels="Fixture Code", axis=1, inplace=True)
```

```python
X = data.to_numpy()
```

```python
# Feature Normalization
scaler = StandardScaler()

y_train = y[:15]
X_train = X[:15]

scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X[15:])

X_train.shape
```

```python
# n_neighbors = ?
neigh = KNeighborsClassifier(n_neighbors=3, weights="distance")
neigh.fit(X_train, y_train)
```

```python
print(neigh.predict(X_test))
print(y[15:])
```

```python
neigh.score(scaler.transform(X), y)
```
